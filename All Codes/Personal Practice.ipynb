```python
class Node:
    def __init__(self, data):
        self.data = data
        self.link = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    # Defaulf Function For Further Use-1
    def insert_first(self, data):
        new_node = Node(data)
        new_node.link = self.head
        self.head = new_node

    # Defaulf Function For Further Use-2
    def insert_last(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        current = self.head
        while current.link is not None:
            current = current.link
        current.link = new_node

    # Defaulf Function For Further Use-3
    def insert_at(self, data, position):
        if position < 0:
            print("Invalid position.")
            return
        elif position == 0:
            self.insert_first(data)
            return
        else:
            new_node = Node(data)
            current = self.head
            current_position = 0
            while current is not None and current_position < position - 1:
                current = current.link
                current_position += 1
            if current is None:
                print("Position exceeds the length of the list. Node not inserted.")
                return
            else:
                new_node.link = current.link
                current.link = new_node

    # Defaulf Function For Further Use-4
    def delete_first(self):
        if self.head is None:
            print("The list is empty. Nothing to delete.")
            return
        temp = self.head
        self.head = self.head.link
        del temp

    # Defaulf Function For Further Use-5
    def delete_last(self):
        if self.head is None:
            print("The list is empty. Nothing to delete.")
            return
        elif self.head.link is None:
            del self.head
            self.head = None
            return
        current = self.head
        while current.link is not None and current.link.link is not None:
            current = current.link
        del current.link
        current.link = None

    # Question-1: Print the size of a linked list.
    def length(self):
      count = 0
      temp = self.head
      while temp is not None:
        count += 1
        temp = temp.next
      return count

    # Question-2: Given the head of a linked list, print the value of the middle node.
    def middle_node(self):
      if self.head is None:
          print("List is empty.")
          return None

      slow = self.head
      fast = self.head

      while fast is not None and fast.link is not None:
          slow = slow.link
          fast = fast.link.link

      print(f"Middle node data: {slow.data}")
      return slow

    # Question-3: Print the number of occurrences of a given number in a linked list.
    def number_of_occurrences(self, num):
      count = 0
      temp = self.head
      while temp is not None:
        if temp.data == num:
          count += 1
        temp = temp.link
      return count

    # Question-4: Given a sorted linked list, insert a number so that after insertion, it remains sorted.
    def insert_sorted(self, num):
      new_node = Node(num)
      if self.head is None or num < self.head.data:
          new_node.link = self.head
          self.head = new_node
          return
      current = self.head
      while current.link is not None and current.link.data < num:
          current = current.link
      new_node.link = current.link
      current.link = new_node
      return


    # Question-5: Delete all the even numbers from a linked list.
    def delete_even_valued_numbers(self):
      current = self.head
      prev = None
      while current is not None:
          if current.data % 2 == 0:
              if prev is not None:
                  prev.link = current.link
              else:
                  self.head = current.link
              del current
              current = prev.link
          else:
              prev = current
              current = current.link
      return

    # Question-7: Reverse a singly linked list.
    def reverse_list(self):
      new_list = SinglyLinkedList()
      current = self.head
      while current is not None:
          new_list.insert_first(current.data)
          current = current.link
      return new_list

    # Question-8: Given a singly linked list and an integer k, rotate the linked list to the left by k places.
    def rotate_list_k_integer(self, k):
      if self.head is None or k == 0:
          return
      tail = self.head
      while tail.link is not None:
        tail = tail.link # last node
      prev = None
      temp = self.head
      for i in range(k):
        prev = temp
        temp = temp.link
      prev.link = None
      tail.link = self.head
      self.head = temp
      return

    # Question-9: Given a singly linked list, print “YES” if it is a palindrome otherwise print “NO”.
    def is_palindrome(self):
      temp = self.head
      other = self.reverse_list().head
      while temp is not None:
        if temp.data != other.data:
          print("NO")
          return False
        temp = temp.link
        other = other.link
      print("YES")
      return True

    # Question-10: Given the head of a linked list, determine if the linked list has a cycle in it.
    def find_if_list_cycling(self):
      slow = self.head
      fast = self.head

      while fast is not None and fast.link is not None:
          slow = slow.link
          fast = fast.link.link

          if slow == fast:
              print("Cycle detected in the list.")
              return True

      print("No cycle found in the list.")
      return False


    def find_value_and_index_cycling_start(self):
      slow = self.head
      fast = self.head

      # Step 1: Detect if a cycle exists
      while fast and fast.link:
          slow = slow.link
          fast = fast.link.link
          if slow == fast:
              break
      else:
          print("No cycle found in the list.")
          return None, -1

      # Step 2: Find the start of the cycle
      slow = self.head
      index = 0
      while slow != fast:
          slow = slow.link
          fast = fast.link
          index += 1

      print(f"Cycle starts at node with value: {slow.data}, index: {index}")
      return slow.data, index


    def delete_item(self, item):
        current = self.head
        prev = None
        while current is not None:
            if current.data == item:
                if prev is not None:
                    prev.link = current.link
                else:
                    self.head = current.link
                del current
                return
            prev = current
            current = current.link
        print("Item not found. Node not deleted.")

    def search_item(self, item):
        current = self.head
        position = 0
        while current is not None:
            if current.data == item:
                return position
            current = current.link
            position += 1
        return -1

    def merge_sorted_lists(self, list2):
        current1 = self.head
        current2 = list2.head
        merged_list = SinglyLinkedList()
        while current1 is not None and current2 is not None:
            if current1.data <= current2.data:
                merged_list.insert_last(current1.data)
                current1 = current1.link
            else:
                merged_list.insert_last(current2.data)
                current2 = current2.link
        while current1 is not None:
            merged_list.insert_last(current1.data)
            current1 = current1.link
        while current2 is not None:
            merged_list.insert_last(current2.data)
            current2 = current2.link
        return merged_list

    def merge_two_lists(self, list2):
        current1 = self.head
        current2 = list2.head
        while current1.link is not None:
            current1 = current1.link
        current1.link = current2
        return self

    def print_list(self):
        current = self.head
        print("Linked list:", end=" ")
        while current is not None:
            print(f"{current.data} ->", end=" ")
            current = current.link
        print("NULL")

def menu():
    linked_list = SinglyLinkedList()

    while True:
        print("\n!!! MENU !!!")
        print("1. Insert First")
        print("2. Insert Last")
        print("3. Insert At")
        print("4. Delete First")
        print("5. Delete Last")
        print("6. Delete Item")
        print("7. Print the list")
        print("8. Search")
        print("9. Quit")
        option = int(input("Select an option: "))

        if option == 1:
            data = int(input("Enter data to insert at the beginning: "))
            linked_list.insert_first(data)
        elif option == 2:
            data = int(input("Enter data to insert at the end: "))
            linked_list.insert_last(data)
        elif option == 3:
            data = int(input("Enter data to insert: "))
            position = int(input("Enter position to insert at: "))
            linked_list.insert_at(data, position)
        elif option == 4:
            linked_list.delete_first()
        elif option == 5:
            linked_list.delete_last()
        elif option == 6:
            item = int(input("Enter the item to delete: "))
            linked_list.delete_item(item)
        elif option == 7:
            linked_list.print_list()
        elif option == 8:
            item = int(input("Enter the item to search: "))
            position = linked_list.search_item(item)
            if position != -1:
                print(f"Item found at position {position}.")
            else:
                print("Item not found in the list.")
        elif option == 9:
            break
        else:
            print("Invalid option. Please select a valid option.")

# Directly call menu() to start the program
menu()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def insert_first(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

    def insert_last(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next is not None:
                current = current.next
            current.next = new_node
            new_node.prev = current
    def delete_last(self):
        if self.head is None:
            print("List is empty. Cannot delete the last node.")
            return
        elif self.head.next is None:
            del self.head
            self.head = None
            return
        else:
            current = self.head
            while current.next is not None:
                current = current.next
            current.prev.next = None
            del current

    # Question-6: Reverse a doubly linked list.
    def reverse_list(self):
        new_list = DoublyLinkedList()
        tail = self.head
        while tail.next is not None:
            new_list.insert_first(tail.data)
            tail = tail.next
        new_list.insert_first(tail.data)
        return new_list
```
