```python
class Node:
  def __init__(self,data):
    self.data=data
    self.next=None
```

```python
class Stack:
  def __init__(self):
    self.top=None
    self.size=0
  def is_empty(self):
    return self.top==None
  def push(self,value):
    new_node=Node(value)
    new_node.next=self.top
    self.top=new_node
    self.size+=1
  def peek(self):
    if self.is_empty():
      return 'Underflow'
    else:
      return self.top.data
  def pop(self):
    if self.is_empty():
      return 'Underflow'
    else:
      data=self.top.data
      self.top=self.top.next
      self.size-=1
      return data
```

```python
class Queue:
  def __init__(self):
    self.front=None
    self.rear=None
    self.size=0
  def is_empty(self):
    return self.front==None
  def enqueue(self,value):
    new_node=Node(value)
    if self.is_empty():
      self.front=new_node
      self.rear=new_node
    else:
      self.rear.next=new_node
      self.rear=new_node
    self.size+=1
  def dequeue(self):
    if self.is_empty():
      return 'Underflow'
    removed_node = self.front
    if self.front==self.rear:
      self.rear=None
      self.front=self.front.next
    else:
      self.front=self.front.next
    self.size-=1
    return removed_node.data

  def peek(self):
    if self.is_empty():
      return 'Underflow'
    else:
      return self.front.data
```

```python
class Graph:
  def __init__(self,vertices):
    self.V=vertices
    self.graph=[]
    for i in range(vertices):
      self.graph.append([])
  def add_vertex(self):
    self.V+=1
    self.graph.append([])
  def add_edge(self,frm,to,bidirectional=True):
    self.graph[frm].append(to)
    if bidirectional:
      self.graph[to].append(frm)
  def print_graph(self):
    for i in range(self.V):
      print(i,end='->')
      for j in self.graph[i]:
        print(j,end=' ')
      print()
  def neighbors(self,vertex):
    return self.graph[vertex]
  def bfs(self,source):
    q=Queue()
    flag=[0]*self.V
    q.enqueue(source)
    flag[source]=1
    while not q.is_empty():
      current_node=q.dequeue()
      neighbors_of_current_node=self.neighbors(current_node)
      for neighbor in neighbors_of_current_node:
        if flag[neighbor]==0:
          q.enqueue(neighbor)
          flag[neighbor]=1
      print(current_node,end=' ')
  def dfs(self,source):
    s=Stack()
    flag=[0]*self.V
    s.push(source)
    flag[source]=1
    while not s.is_empty():
      current_node=s.pop()
      neighbors_of_current_node=self.neighbors(current_node)
      for neighbor in neighbors_of_current_node:
        if flag[neighbor]==0:
          s.push(neighbor)
          flag[neighbor]=1
      print(current_node,end=' ')
  def indegree(self,vertex):
    count=0
    for i in range(self.V):
      if vertex in self.graph[i]:
        count+=1
    return count
  def outdegree(self,vertex):
    return len(self.graph[vertex])
```

Every vertex pair is connected by at most one edge, and no vertex has an edge to itself.

You want to determine if there is a valid path that exists from vertex source to vertex destination.

Given edges and the integers n, source, and destination, return true if there is a valid path from source to destination, or false otherwise.

If the town judge exists, then:

The town judge trusts nobody.
Everybody (except for the town judge) trusts the town judge.
There is exactly one person that satisfies properties 1 and 2.
You are given an array trust where trust[i] = [ai, bi] representing that the person labeled ai trusts the person labeled bi. If a trust relationship does not exist in trust array, then such a trust relationship does not exist.

Return the label of the town judge if the town judge exists and can be identified, or return -1 otherwise.



Example 1:

Example 2:

Input: n = 3, trust = [[1,3],[2,3]]

```python
class GGraph:
  def __init__(self,vertices):
    self.V=vertices
    self.graph=[]
    for i in enumerate(range(vertices),start=1):
      self.graph.append([])
  def add_vertex(self):
    self.V+=1
    self.graph.append([])
  def add_edge(self,frm,to,bidirectional=True):
    self.graph[frm-1].append(to)
    if bidirectional:
      self.graph[to-1].append(frm)
  def print_graph(self):
    for i in range(self.V):
      print(i+1,end='->')
      for j in self.graph[i]:
        print(j,end=' ')
      print()
  def neighbors(self,vertex):
    return self.graph[vertex-1]
  def bfs(self,source):
    q=Queue()
    flag=[0]*self.V
    q.enqueue(source)
    flag[source]=1
    while not q.is_empty():
      current_node=q.dequeue()
      neighbors_of_current_node=self.neighbors(current_node)
      for neighbor in neighbors_of_current_node:
        if flag[neighbor]==0:
          q.enqueue(neighbor)
          flag[neighbor]=1
      print(current_node,end=' ')
  def dfs(self,source):
    s=Stack()
    flag=[0]*self.V
    s.push(source)
    flag[source]=1
    while not s.is_empty():
      current_node=s.pop()
      neighbors_of_current_node=self.neighbors(current_node)
      for neighbor in neighbors_of_current_node:
        if flag[neighbor]==0:
          s.push(neighbor)
          flag[neighbor]=1
      print(current_node,end=' ')
  def indegree(self,vertex):
    count=0
    for i in range(self.V):
      if vertex in self.graph[i]:
        count+=1
    return count
  def outdegree(self,vertex):
    return len(self.graph[vertex-1])
```

```python
def findJudge(self):
  for i in range(1,self.V+1):
    if self.outdegree(i)==0 and self.indegree(i)==self.V-1:
      return i
  return -1
```

There is an undirected star graph consisting of n nodes labeled from 1 to n. A star graph is a graph where there is one center node and exactly n - 1 edges that connect the center node with every other node.

You are given a 2D integer array edges where each edges[i] = [ui, vi] indicates that there is an edge between the nodes ui and vi. Return the center of the given star graph.



Example 1:


Input: edges = [[1,2],[2,3],[4,2]]

Explanation: As shown in the figure above, node 2 is connected to every other node, so 2 is the center.
Example 2:

######Q-Number of Perimeter of Island(LeetCode)

```python
def islandPerimeter(grid):
  height=len(grid)
  width=len(grid[0])
  def dfs(r,c):
    if r<0 or r>=height or c<0 or c>=width or grid[r][c]==0:
      return 1
    if grid[r][c]==1:
      grid[r][c]=2
      return dfs(r+1,c)+dfs(r-1,c)+dfs(r,c+1)+dfs(r,c-1)
    return 0
  perimeter=0
  for r in range(height):
    for c in range(width):
      if grid[r][c]==1:
        perimeter+=dfs(r,c)
  return perimeter

```

######Q-Check if a Graph is Bipartite(Using BFS)

```python
def check_bipartite(self):
  color=[-1]*self.V
  for start in range(self.V):
    if color[start]==-1:
      q=Queue()
      q.enqueue(start)
      color[start]=0
      while not q.is_empty():
        current_node=q.dequeue()
        for neighbor in self.neighbors(current_node):
          if color[neighbor]==-1:
            color[neighbor]=1-color[current_node]
            q.enqueue(neighbor)
          elif color[neighbor]==color[current_node]:
            return False
  return True
```

######Q-Detect a Cycle in a Graph Using BFS

```python
def is_cyclic(self,source):
  flag=[-1]*self.V
  q=Stack()
  q.push(source)
  flag[source]=0
  while not q.is_empty():
    current_node=q.pop()
    flag[current_node]=1
    neighbors_of_current_node=self.graph[current_node]
    for neighbor in neighbors_of_current_node:
      if flag[neighbor]==-1:
        q.push(neighbor)
      elif flag[neighbor]==1:
        return True
  return False
```

```python
def is_connected(self,source):
  flag=[0]*self.V
  s=Stack()
  s.push(source)
  flag[source]=1
  while not s.is_empty():
    current_node=s.pop()
    neighbors_of_current_node=self.graph[current_node]
    for neighbor in neighbors_of_current_node:
      if flag[neighbor]==0:
        s.push(neighbor)
        flag[neighbor]=1
  for i in range(self.V):
    if flag[i]==0:
      return False
  return True
```

```python
def find_connected_components(self):
  compo=0
  flag=[0]*self.V
  for i in range(self.V):
    if flag[i]==0:
      compo+=1
      s=Stack()
      s.push(i)
      flag[i]=1
      while not s.is_empty():
        current_node=s.pop()
        neighbors_of_current_node=self.graph[current_node]
        for neighbor in neighbors_of_current_node:
          if flag[neighbor]==0:
            s.push(neighbor)
            flag[neighbor]=1
  return compo
```
