## 1. Two Sum (LC - 1)

```python
def two_sum(nums, target):
  for i in range(len(nums)-1):
    for j  in range(i+1, len(nums)):
      if nums[i]+nums[j] == target:
        return [i,j]
  return -1

nums = [3,2,4]
target = 6
two_sum(nums, target)
```

```python
def find_the_distance_value_between_two_arrays(arr1, arr2, d):
  count = 0
  for i in arr1:
    for j in arr2:
      if abs(i-j) <= d:
        break
    else:
      count += 1
  return count
arr1 = [4,5,8]
arr2 = [10,9,1,8]
d = 2
find_the_distance_value_between_two_arrays(arr1, arr2, d)
```

## 4. Add Two Numbers (LC - 2)

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.link = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def insert_first(self, data):
        new_node = Node(data)
        new_node.link = self.head
        self.head = new_node

    def insert_last(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        current = self.head
        while current.link is not None:
            current = current.link
        current.link = new_node

    def show_list(self):
        current = self.head
        while current is not None:
            print(current.data, end=" -> ")
            current = current.link
        print("None")

    def reverse_list(self):
      new_list = SinglyLinkedList()
      current = self.head
      while current is not None:
          new_list.insert_first(current.data)
          current = current.link
      return new_list

def add_two_numbers(l1, l2):  # l1 and l2 must be SinglyLinkedList
    l3 = SinglyLinkedList()
    carry = 0

    p1 = l1.reverse_list().head
    p2 = l2.reverse_list().head

    while p1 is not None or p2 is not None:
        l1_data = p1.data if p1 is not None else 0
        l2_data = p2.data if p2 is not None else 0

        total = l1_data + l2_data + carry  # include carry
        carry = total // 10
        digit = total % 10

        l3.insert_last(digit)

        if p1 is not None:
            p1 = p1.link
        if p2 is not None:
            p2 = p2.link

    # After the loop, if carry remains, add it
    if carry > 0:
        l3.insert_last(carry)

    l3.reverse_list().show_list()  # You can return l3 instead if you want

l1 = SinglyLinkedList()
l1.insert_last(1)
l1.insert_last(8)
# l1.insert_last(3)

l2 = SinglyLinkedList()
l2.insert_last(1)
l2.insert_last(7)
# l2.insert_last(4)

add_two_numbers(l1, l2)
```

## 5. Relative Sort Array (LC - 1122)

```python
def relative_sort_array(arr1, arr2):
  result = []
  for i in arr2:
    for j in arr1:
      if i == j:
        result.append(j)

  extra_element_lst = [i for i in arr1 if i not in arr2]
  n = len(extra_element_lst)
  for i in range(n - 1):
    for j in range(n - 1 - i):
      if extra_element_lst[j] > extra_element_lst[j + 1]:
          extra_element_lst[j], extra_element_lst[j + 1] = extra_element_lst[j + 1], extra_element_lst[j]

  for i in extra_element_lst:
    if i not in result:
      result.append(i)

  return result

arr1 = [28,6,22,8,44,17]
arr2 = [22,28,8,6]
relative_sort_array(arr1, arr2)
```

## 1. Valid Parentheses (LC - 20)

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Stack:
    def __init__(self):
        self.head = None
    def isEmpty(self):
        return True if self.head == None else False
    def push(self, data):
        if self.head == None:
            self.head = Node(data)
        else:
            temp = Node(data)
            temp.next = self.head
            self.head = temp
    def pop(self):
        if self.head == None:
            return "[ERROR] Stack Underflow!"
        else:
            temp = self.head
            self.head = self.head.next
            temp.next = None
            data = temp.data
            del temp
            return data
    def peek(self):
      if self.head == None:
          return None
      return self.head.data
```

```python
def valid_parenthesis(string):
    s = Stack()
    for char in string:
        if char in "({[":
            s.push(char)
        elif char in ")}]":
            if s.isEmpty():
                return False
            top = s.peek()
            if char == ")" and top == "(":
                s.pop()
            elif char == "}" and top == "{":
                s.pop()
            elif char == "]" and top == "[":
                s.pop()
            else:
                return False
    return s.isEmpty()
valid_parenthesis('[()]')
```

##3. Baseball Game (LC - 682)

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Stack:
    def __init__(self):
        self.head = None
    def isEmpty(self):
        return True if self.head == None else False
    def push(self, data):      # Link List er insert first hoitese.
        if self.head == None:
            self.head = Node(data)
        else:
            temp = Node(data)
            temp.next = self.head
            self.head = temp
    def pop(self):          # Tai pop er time e just head theke remove korlei hobe.
        if self.head == None:
            return "[ERROR] Stack Underflow!"
        else:
            temp = self.head
            self.head = self.head.next
            temp.next = None
            data = temp.data
            del temp
            return data
    def peek(self):
      if self.head == None:
          return None
      return self.head.data
    def len(self):
      count = 0
      current = self.head
      while current is not None:
        count += 1
        current = current.next
      return count
```

```python
def baseball_game(ops):
  stack = Stack()
  for i in ops:
    if i == '+':
      if stack.isEmpty():
        return "Not yet scored!"
      else:
        data1 = stack.pop()
        data2 =  stack.pop()
        total = int(data1) + int(data2)
        stack.push(data2)
        stack.push(data1)
        stack.push(total)
    elif i == 'D':
      if stack.isEmpty():
        return "Not yet scored!"
      else:
        data = stack.pop()
        total = int(data) * 2
        stack.push(data)
        stack.push(total)
    elif i == 'C':
      if stack.isEmpty():
        return "Not yet scored!"
      else:
        stack.pop()
    else:
      stack.push(i)
  sum_of_all_run = 0
  while not stack.isEmpty():
    sum_of_all_run += int(stack.pop())
  return sum_of_all_run

ops = ["5","2","C","D","+"]
print(baseball_game(ops))
ops = ["5","-2","4","C","D","9","+","+"]
print(baseball_game(ops))
ops = ["1","C"]
print(baseball_game(ops))
```

```python
def backspace_string_compare(s,t):
  stack_s = Stack()
  stack_t = Stack()
  for i in s:
    if i == '#':
      if stack_s.isEmpty():
        continue
      else:
        stack_s.pop()
    else:
      stack_s.push(i)
  for i in t:
    if i == '#':
      if stack_t.isEmpty():
        continue
      else:
        stack_t.pop()
    else:
      stack_t.push(i)
  while not stack_s.isEmpty() and not stack_t.isEmpty():
    if stack_s.pop() != stack_t.pop():
      return False
  return stack_s.isEmpty() and stack_t.isEmpty()

s = "ab#c"
t = "ad#c"
print(backspace_string_compare(s,t))

s = "ab##"
t = "c#d#"
print(backspace_string_compare(s,t))

s = "a#c"
t = "b"
print(backspace_string_compare(s,t))
```

```python
class Queue:
  def __init__(self):
    self.head = None
    self.tail = None

  def enqueue(self, data):
    if self.head == None:
      self.head = Node(data)
      self.tail = self.head
    else:
      # self.tail.next = Node(data)
      # self.tail = self.tail.next
      temp = Node(data)
      self.tail.next = temp
      self.tail = temp

  def dequeue(self):
    if self.head == None:
      print('[Error] Queue Underflow')
      return None
    else:
      temp = self.head
      self.head = self.head.next
      temp.next = None
      data = temp.data
      del temp
      return data

  def show_queue(self):
    temp = self.head
    while temp != None:
      if temp.next != None:
        print(temp.data, end='-->')
        temp = temp.next
      else:
        print(temp.data)
        temp = temp.next
    print()

  def front_element_return(self):
    if self.head == None:
      print('[Error] Queue Underflow')
      return None
    else:
      return self.head.data

  def isEmpty(self):
    return self.head == None

  def len(self):
    temp = self.head
    count = 0
    while temp != None:
      count += 1
      temp = temp.next
    return count
```

```python
def first_unique_character_in_a_string(s):
  queue = Queue()
  freq_dict = {}
  for i in s:
    count = 0
    for j in s:
      if i == j:
        count += 1
    freq_dict[i] = count
  index = 0
  for i in s:
    if freq_dict[i] == 1:
      queue.enqueue((i, index))
    index += 1
  if queue.isEmpty():
    return -1
  else:
    return queue.dequeue()[1]

s = "leetcode"
print(first_unique_character_in_a_string(s))

s = "loveleetcode"
print(first_unique_character_in_a_string(s))

s = "aabb"
print(first_unique_character_in_a_string(s))
```

##3. Number of Students Unable to Eat Lunch (LC - 1700):

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Stack:
    def __init__(self):
        self.head = None
    def isEmpty(self):
        return True if self.head == None else False
    def push(self, data):
        if self.head == None:
            self.head = Node(data)
        else:
            temp = Node(data)
            temp.next = self.head
            self.head = temp
    def pop(self):
        if self.head == None:
            return "[ERROR] Stack Underflow!"
        else:
            temp = self.head
            self.head = self.head.next
            temp.next = None
            data = temp.data
            del temp
            return data
    def peek(self):
      if self.head == None:
          return None
      return self.head.data

    def show_stack(self):
        temp = self.head
        while temp != None:
            if temp.next != None:
                print(temp.data, end='-->')
                temp = temp.next
            else:
                print(temp.data)
```

```python
def number_of_students_unable_to_eat_lunch(students, sandwiches):
    queue = Queue()
    for i in students:
        queue.enqueue(i)

    stack = Stack()
    for i in reversed(sandwiches):  # reverse since sandwiches[0] is the top
        stack.push(i)

    no_eat_count = 0

    while not queue.isEmpty() and no_eat_count < queue.len():
        student = queue.dequeue()
        top_sandwich = stack.peek()

        if student == top_sandwich:
            stack.pop()
            no_eat_count = 0  # reset if someone eats
        else:
            queue.enqueue(student)  # move student to end
            no_eat_count += 1       # one more refused

    return queue.len()

students = [1,1,0,0]
sandwiches = [0,1,0,1]
print(number_of_students_unable_to_eat_lunch(students, sandwiches))

students = [1,1,1,0,0,1]
sandwiches = [1,0,0,0,1,1]
print(number_of_students_unable_to_eat_lunch(students, sandwiches))
```

##4. Time Needed to Buy Tickets (LC - 2073)

```python
def time_needed_to_buy_tickets(tickets, k):
    queue = Queue()
    for i in range(len(tickets)):
        queue.enqueue((tickets[i], i))  # (tickets_left, original_index)

    time = 0

    while True:
        person = queue.dequeue()   # It's a tuple
        tickets_left, idx = person

        # Simulate 1 second of buying ticket
        tickets_left -= 1
        time += 1

        if idx == k and tickets_left == 0:
            return time

        if tickets_left > 0:
            queue.enqueue((tickets_left, idx))
    return time

tickets = [2,3,2]
k = 2
print(time_needed_to_buy_tickets(tickets, k))

tickets = [5,1,1,1]
k = 0
print(time_needed_to_buy_tickets(tickets, k))
```

## Shortest path using BFS

image.png

```python
class Graph:
    def __init__(self, vertices):
        self.V = vertices
        self.graph = []
        for i in range(vertices):
            self.graph.append([])


    def add_edge(self, frm, to, bidirectional=True):
        self.graph[frm].append(to)
        if bidirectional == True:
            self.graph[to].append(frm)

    def print_graph(self):
        for i in range(self.V):
            print(f"{i} -> ", end="")
            for j in self.graph[i]:
                print(f"{j} ", end="")
            print()

    def neighbors(self, vertex):
        return self.graph[vertex]

    def bfs(self, source):
        queue = Queue()
        flag = [0] * self.V

        queue.enqueue(source)
        flag[source] = 1

        while not queue.isEmpty():
            current_node = queue.dequeue()

            print(current_node, end=" ")

            neighbors_of_current_node = self.neighbors(current_node)
            for neighbor in neighbors_of_current_node:
                if flag[neighbor] == 0:
                    queue.enqueue(neighbor)
                    flag[neighbor] = 1

    def bfs_shortest_path(self, source):
        queue = Queue()
        visited = [False] * self.V
        distance = [float('inf')] * self.V
        parent = [-1] * self.V  # Optional: for path reconstruction

        queue.enqueue(source)
        visited[source] = True
        distance[source] = 0

        while not queue.isEmpty():
            current = queue.dequeue()

            for neighbor in self.neighbors(current):
                if not visited[neighbor]:
                    queue.enqueue(neighbor)
                    visited[neighbor] = True
                    distance[neighbor] = distance[current] + 1
                    parent[neighbor] = current

        # Print shortest distances from source
        print("Node\tDistance from Source")
        for i in range(self.V):
            print(f"{i}\t{distance[i]}")

        # Optional: return distance or path
        return distance

    def bfs_shortest_path_to_target(self, source, target):
      queue = Queue()
      visited = [False] * self.V
      parent = [-1] * self.V

      queue.enqueue(source)
      visited[source] = True

      while not queue.isEmpty():
          current = queue.dequeue()

          if current == target:
              break

          for neighbor in self.neighbors(current):
              if not visited[neighbor]:
                  queue.enqueue(neighbor)
                  visited[neighbor] = True
                  parent[neighbor] = current

      # Reconstruct path from target to source using parent[]
      path = []
      temp = target
      while temp != -1:
          path.append(temp)
          temp = parent[temp]

      # If the target is unreachable from source
      if path[-1] != source:
          print("No path found.")
          return []

      path.reverse()  # Make it source -> target
      return path
```

##1. Island Perimeter (LC - 463)

```python
class Graph:
    def __init__(self, vertices):
        self.V = vertices
        self.graph = []
        for i in range(vertices):
            self.graph.append([])


    def add_edge(self, frm, to, bidirectional=True):
        self.graph[frm].append(to)
        if bidirectional == True:
            self.graph[to].append(frm)

    def print_graph(self):
        for i in range(self.V):
            print(f"{i} -> ", end="")
            for j in self.graph[i]:
                print(f"{j} ", end="")
            print()

    def neighbors(self, vertex):
        return self.graph[vertex]

    def island_perimeter(self, grid):
        rows = len(grid)
        cols = len(grid[0])
        perimeter = 0

        for r in range(rows):
          for c in range(cols):
            if grid[r][c] == 1:
              perimeter += 4

              # Check up
              if r > 0 and grid[r - 1][c] == 1:
                perimeter -= 1
              # Check down
              if r < rows - 1 and grid[r + 1][c] == 1:
                perimeter -= 1
              # Check left
              if c > 0 and grid[r][c - 1] == 1:
                perimeter -= 1
              # Check right
              if c < cols - 1 and grid[r][c + 1] == 1:
                perimeter -= 1
        return perimeter
```

##3. Find the Town Judge (LC - 997)

```python
def findJudge(n, trust):
    in_degree = [0] * (n + 1)
    out_degree = [0] * (n + 1)

    for a, b in trust:
        out_degree[a] += 1  # a trusts someone
        in_degree[b] += 1   # b is trusted by someone

    for i in range(1, n + 1):
        if in_degree[i] == n - 1 and out_degree[i] == 0:
            return i  # i is the judge

    return -1  # no judge found
```

##4. Find Center of Star Graph (LC - 1791)

```python
class Graph:
    def __init__(self, vertices):
        self.V = vertices
        self.graph = []
        for i in range(vertices):
            self.graph.append([])


    def add_edge(self, frm, to, bidirectional=True):
        self.graph[frm].append(to)
        if bidirectional == True:
            self.graph[to].append(frm)

    def print_graph(self):
        for i in range(self.V):
            print(f"{i} -> ", end="")
            for j in self.graph[i]:
                print(f"{j} ", end="")
            print()

    def neighbors(self, vertex):
        return self.graph[vertex]

    def in_degree(self):
        in_degree = [0] * self.V
        for i in range(self.V):
            for j in self.graph[i]:
                in_degree[j] += 1
        return in_degree

    def out_degree(self):
        out_degree = [0] * self.V
        for i in range(self.V):
            out_degree[i] = len(self.graph[i])
        return out_degree

    def find_center_of_star_graph(self):
        in_degree = self.in_degree()
        out_degree = self.out_degree()
        for i in range(self.V):
            if in_degree[i] == self.V - 1 and out_degree[i] == self.V - 1:
                return i
        return -1

graph = Graph(4)
graph.add_edge(0, 2)
graph.add_edge(2, 1)
graph.add_edge(3, 2)
print(graph.find_center_of_star_graph())
```
